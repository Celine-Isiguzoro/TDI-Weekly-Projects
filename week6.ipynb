Task 4: Create New Columns Through Calculations Create derived columns that combine existing data. 
Write code to: 
•	Create a column for order_profit (cost * quantity * 0.25 for 25% margin) 
•	Create a column for efficiency_score (lead_time / actual_delivery_days) 
•	Create a column for high_value_order (True if cost > average_cost, False otherwise) 
•	Create a column for delay_status ("On Time" or "Delayed" based on actual vs expected delivery) 
•	Display the first 10 rows with these new columns 

```python
# Create a column for order_profit (cost * quantity * 0.25 for 25% margin) 
df['order_profit'] = df['total_cost'] * df['quantity'] * 0.25
df['order_profit'].head().round(2)
```

```python
# Create a column for efficiency_score (lead_time / actual_delivery_days)
df['efficiency_score'] = df['lead_time_days'] / df['actual_delivery_days']
df['efficiency_score'].head()
```

```python
# Create a column for high_value_order (True if cost > average_cost, False otherwise)
df['high_value_order'] = df['total_cost'] >  df['total_cost'].mean()
df['high_value_order'].head()
```

```python
# Create a column for delay_status ("On Time" or "Delayed" based on actual vs expected delivery)
import numpy as np

df['delay_status'] = np.where(df['actual_delivery_days'] <= df['lead_time_days'], 'On_time', 'Delayed')
df['delay_status'].head()
```

Task 5: Merge with Supplier Reference Table 
A supplier reference table (supplier_reference.csv) contains contact and location information for each supplier. Load this table and merge it with your orders data. 
Write code to: 
•	Load the supplier reference table 
•	Perform a merge with your orders DataFrame on supplier_name 
•	Display the merged data with both order information and supplier details 
•	Verify that all suppliers matched correctly 
•	Show what new columns were added from the merge 

```python
df_ref = pd.read_csv('supplier_reference.csv')
df_ref.head()
```

```python
# Perform a merge with your orders DataFrame on supplier_name 
# Standardize the supplier names on both tables
df['supplier_name'] = df['supplier_name'].str.strip().str.replace(' ', '').str.title()
df.supplier_name.unique()

supplier_naming = {
    'Suppliera': 'SupplierA',
    'Supplierb': 'SupplierB',
    'Supplierc': 'SupplierC',
    'Supplierd': 'SupplierD',
    'Suppliere': 'SupplierE'
}
df['supplier_name'] = df['supplier_name'].replace(supplier_naming)
df.supplier_name.unique()

# Merge columns
df_merge = df.merge(df_ref, on='supplier_name', how='left')
df_merge.head()
```

```python
new_cols = []
for col in df_merge.columns:
    if col not in df.columns:
        new_cols.append(col)
print(new_cols)
```

Challenge 1: 
Create a pivot table showing average order cost by supplier and product category. What patterns emerge? 

```python
pivot_table = df_merge.pivot_table(values='total_cost',
                                   index='supplier_name',
                                   columns='product_category',
                                   aggfunc='mean').round(2)
pivot_table.head()
```

1. All suppliers are specialized i.e no supplier covers all product categories
2. SupplierB has the highest average cost with Motors category
3. SupplierC has the least average cost with Fasteners category

Challenge 2: 
Find the top 10 customers (by total spending). Create a DataFrame with customer_id, total_spending, order_count, and average_order_value. 

```python
df_merge.columns
```

```python
top_ten_customers = df_merge.groupby('order_id').agg(
    total_spending = ('total_cost', 'sum'),
    order_count = ('order_id', 'count'),
    average_order_value = ('total_cost', 'mean')
).round(2).nlargest(10, 'total_spending').reset_index()

top_ten_customers
```

Challenge 3: 
Use .groupby() to calculate statistics by supplier: average lead time, average delivery time, on-time delivery rate, total orders. 
Which supplier performs best overall? 

```python
stat_by_supp = df_merge.groupby('supplier_name').agg(
    avg_delivery_time = ('actual_delivery_days', 'mean'),
    avg_lead_time = ('lead_time_days', 'mean'),
    on_time_del_rate = ('on_time', lambda z: (z== 'Yes').mean() * 100),
    total_orders = ('order_id', 'count')
).round(2).nlargest(5, 'total_orders')

stat_by_supp

# SupplierA performs the best because of the following reasons:
# Highest orders(769)
# Fastest average delivery time(10.51 days)
# Highest on time delivery rate(94.28%)
# Shortest average lead time(9.94 days) 
```
