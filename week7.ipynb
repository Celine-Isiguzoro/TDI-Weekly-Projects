```python
import pandas as pd
import numpy as np

df = pd.read_csv("supply_chain_dataset.csv")

# To know the count of missing values for each column
df_isnull = df.isnull().sum()
# print(df_isnull)

# Dropping the missing values because there are many and return status could be misleading
df.dropna(subset=['product_defects', 'return_status'], inplace=True)

# View missing values
print(df.isnull().sum())
```

```python
# Check the data types of each column
print(f"The data types in the dataset are: \n{df.dtypes}")

# Change the data type of the date columns
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed')
df['delivery_date'] = pd.to_datetime(df['delivery_date'], format='mixed')

# To view
print(f"The converted data type for delivery_date column is {df.delivery_date.dtype}")
print(f"The converted data type for order_date column is {df.order_date.dtype}")

# Changing numeric text to actual numbers
# All numerical columns were stored correctly but let's use one to set an example
df['quantity'] = pd.to_numeric(df['quantity'])

# To confirm all data types
print(df.dtypes)
```

Task 3: Create Simple Calculated Columns 
Create new columns by combining or transforming existing data. 
Create these new columns: Column 1: Days to Deliver 
•	Subtract order_date from delivery_date to get number of days 
•	This tells you how long each delivery took Column 2: Order Size Category 
•	If quantity < 50: "Small" 
•	If quantity 50-200: "Medium" 
•	If quantity > 200: "Large" 
•	This groups orders by size Column 3: High Cost Flag 
•	If total_cost > average total_cost: 1 (True) 
•	Otherwise: 0 (False) 
•	This marks expensive orders 
Write code to create all three columns. Display sample data showing the new columns


```python
# Create days_to_deliver column
df['days_to_deliver'] = df['delivery_date'] - df['order_date']
df['days_to_deliver'].dt.days

# Create order_size_category column
def order_size_category(quantity):
    if quantity < 50:
        return 'Small'
    elif quantity <= 200:
        return 'Medium'
    else:
        return 'Large'

df['order_size_category'] = df['quantity'].apply(order_size_category)
df['order_size_category']

# Create high_cost_flag column (To mark expensive orders)
df['high_cost_flag'] = (df['total_cost'] > df['total_cost'].mean()).astype(int)
df['high_cost_flag']

# Display new columns
df[['order_date', 'delivery_date', 'days_to_deliver', 'order_size_category', 'quantity', 'high_cost_flag', 'total_cost' ]]
```

Task 4: Clean Up Text Data 
Text data is often messy. Clean it so it's consistent. 
Write code to:

•	Make all supplier_name values the same capitalization (all uppercase or title case) 
•	Check for spelling inconsistencies (like "Motor" vs "Moter") and fix any you find

```python
# Removing extra spaces from supplier_name and product_category
print(f'Before cleaning: {df.head(10)}')
# df[['supplier_name', 'product_category']].head(10)

df['supplier_name'] = df['supplier_name'].str.strip()
df['supplier_name'] = df['supplier_name'].str.replace(' ','')
df['product_category'] = df['product_category'].str.strip()

# Standardization
df['supplier_name'] = df['supplier_name'].str.title()
df['product_category'] = df['product_category'].str.title()

# Checking inconsistencies
df['supplier_name'].values.unique()
df['product_category'].values.unique()

# After cleaning
print('After cleaning: ')
df.head(15)

```

Task 5: Create Performance Metrics 
Create columns that show how well things performed. 
Create these columns: 
Column 1: On Time (1 or 0) 
•	If days_to_deliver <= lead_time_days: 1 (on time) 
•	Otherwise: 0 (late) 
•	This flags the orders that arrived on schedule Column 2: Cost Per Unit 
•	Divide total_cost by quantity 
•	This shows the unit price 
•	Helps compare if one order was more expensive per item 
Column 3: Supplier Performance Grade 
•	For each supplier, calculate their on-time delivery rate 
•	Merge this back to each order 
•	This shows each supplier's overall reliability 
Write code to create all three. Display orders with their performance metrics. 

```python
# Extract days from the time delta values for days_to_deliver
# df['days_to_deliver'] = df['days_to_deliver'].dt.days

# Create an on_time_num column
df['on_time_num'] = (df['days_to_deliver'] <= df['lead_time_days']).astype(int)

# Create a cost_per_unit column
df['cost_per_unit'] = df['total_cost'] / df['quantity']

# Create a supplier_perf_grade column
supp_perf =  df.groupby('supplier_name')['on_time_num'].mean().reset_index()
supp_perf.columns = ['supplier_name', 'supplier_on_time_num_rate']
# Merge columns
df = df.merge(supp_perf, on='supplier_name', how='left')
# df.drop('supplier_on_time_rate')
df.head(10)
```

Challenge 1: 
Create a "Delivery Speed Category" column: "Fast" (< 7 days), "Normal" (7-21 days), "Slow" (> 21 days). How many orders fall into each category? 

```python
# Create a delivery_speed_cat column
def del_speed_cat(days):
    if days < 7:
        return 'Fast'
    elif days <= 21:
        return 'Normal'
    else:
        return 'Slow'
df['delivery_speed_cat'] = df['days_to_deliver'].apply(del_speed_cat)
df['delivery_speed_cat'].value_counts().unique
```

Challenge 3: 
For each product_category, calculate the average cost. Then create a column showing if each order's cost is above or below the category average. 

```python
# Compare average cost and product_category
category_avg = df.groupby('product_category')['total_cost'].transform('mean')

# Attach categorical data to ascertain whether total cost is above or below category average
df['category_avg'] = np.where(df['total_cost'] > category_avg, 'Above', 'Below')

# Display the columns
df[['total_cost', 'product_category', 'category_avg']]
```
